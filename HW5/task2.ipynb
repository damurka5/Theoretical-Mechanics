```python
from math import *
import numpy as np
import matplotlib.pyplot as plt

m = 0.02
w = pi
r = 0.5
t_start = 0
t_end = 0.388972431
x0 = 0
v0 = 0.4
g = 9.81


dt = 0.001
n_steps = int(round((t_end-t_start)/dt))

c2 = (-0.4-(g/(2*w)))*(1/(2*w))
c1 = -c2
x_t = []
t_arr = np.linspace(t_start, t_end, n_steps)
N_t = []

for i in range(0, n_steps):
    x_t.append(c1*e**(w*t_arr[i]) + c2*e**(-w*t_arr[i]) - (g/(2*w**2))*sin(w*t_arr[i]))
    v_r = w*(c1*e**(w*t_arr[i]) - c2*e**(-w*t_arr[i])) - (g/(2*w))*cos(w*t_arr[i])
    N_t.append(2*m*v_r*w-m*g*cos(w*t_arr[i]))
```

```python
# plotting the result
fig = plt.figure(figsize=(15,10))                                  # create figure
plt.plot(t_arr, N_t, linewidth = 1, label = 'N(t)')
plt.title('N(t)', fontsize = 12)    # add some title to your plot
plt.xlabel('t (in seconds)', fontsize = 12)
plt.ylabel('N(t)', fontsize = 12)
plt.grid(True)                        # show grid
plt.axis([t_start-0.1, t_end+0.1, -0.2, 0.4])     # show axes measures
plt.legend()
plt.savefig('task2n.png', dpi=300)
plt.show()
```
